```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python

iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
print("Feature Names: ", feature_names)
print("Target Names: ", target_names)
print("10 Rows: \n", X[:10])
print("10 Rows: \n", y[:10])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.4, random_state = 1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
clasif_knn = KNeighborsClassifier(n_neighbors = 3)
clasif_knn.fit(X_train, y_train)
y_pred = clasif_knn.predict(X_test)
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))


sample = [[5, 5, 3, 2], [2, 4, 3, 5]]

preds = clasif_knn.predict(sample)
pred_species = [iris.target_names[p] for p in preds] 
print("Predictions: ", pred_species)
```

```python
import numpy as np
from sklearn import preprocessing

ip_data = np.array([[1.2, 2.1, -2.0], 
                   [0.4, 3.4, -1.4],
                   [-5.6, 5.0, 7.8]])
data_binarised = preprocessing.Binarizer(threshold = 0.5).transform(ip_data)
print(data_binarised)
print(ip_data.mean(axis = 0))
print(ip_data.std(axis = 0))

data_scaled = preprocessing.scale(ip_data)
print(data_scaled.mean(axis = 0))
print(data_scaled.std(axis = 0))
```
